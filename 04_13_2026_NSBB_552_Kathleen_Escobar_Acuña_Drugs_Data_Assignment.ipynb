```python
import pandas as pd
from scipy import stats

# 1. Load your data
# Make sure the file name matches exactly what you see in your folder
df = pd.read_csv('drugs-data.csv') 

# 2. Check for Normality (Shapiro-Wilk Test)
# Replace 'Group1' and 'Group2' with your actual column names
group1 = df.iloc[:, 0] # This selects the first column
group2 = df.iloc[:, 1] # This selects the second column

shapiro_g1 = stats.shapiro(group1)
shapiro_g2 = stats.shapiro(group2)

print(f"Shapiro-Wilk Test Group 1: p-value = {shapiro_g1.pvalue:.4f}")
print(f"Shapiro-Wilk Test Group 2: p-value = {shapiro_g2.pvalue:.4f}")

# 3. Check for Homogeneity of Variances (Levene's Test)
levene_test = stats.levene(group1, group2)
print(f"Levene's Test: p-value = {levene_test.pvalue:.4f}")

# 4. Decision Logic
alpha = 0.05
if shapiro_g1.pvalue > alpha and shapiro_g2.pvalue > alpha and levene_test.pvalue > alpha:
    print("\nRESULT: Data is PARAMETRIC.")
    print("Action: You should use a T-test or ANOVA.")
else:
    print("\nRESULT: Data is NON-PARAMETRIC.")
    print("Action: You should use Mann-Whitney U or Kruskal-Wallis.")
```
